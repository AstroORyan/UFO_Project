# Getting Data From PDF
All of the information from the UK government on alien encounters is stored in PDF tabulated data. Need to extract this and put it into something usable.

## Imports

```python
import glob
import os
import sys
import numpy as np
from tqdm.notebook import tqdm
from tabula import read_pdf

import pandas as pd
```

## Getting the PDF data

```python
data_dir = 'C:/Users/dorya/Documents/UFO_Project/data/pdfs'
save_dir = 'C:/Users/dorya/Documents/UFO_Project/data/csvs'
```

```python
pdf_files = glob.glob(f'{data_dir}/*.pdf')
pdf_file = pdf_files[0]
print(pdf_file)
```

```python
dfs = read_pdf(pdf_file, pages = 'all')
```

```python
dfs[0].head()
```

```python
df = (
    dfs
    [0]
    .drop(columns = ['Town / Village'])
    .rename(columns = {
        'Occupation of reporter' : 'Occupation of reporter (where known)',
        'Area' : 'County',
        'Unnamed: 0' : 'Town / Village', 
    })
    [1:]
)
```

```python
df
```

```python
for i, df_tmp in enumerate(dfs):
    if i == 0:
        continue
    data = np.array(df_tmp)
    
    df_tmp = pd.DataFrame(data=data, columns = range(0,len(data[0])))
    
    df_rename = (
        df_tmp
        .rename(
            columns = {
                0 : 'Date',
                1 : 'Time',
                2: 'Town / Village',
                3 : 'County',
                4: 'Occupation of reporter (where known)',
                5 : 'Brief Description of sighting'
            }
        )
    )

    if 6 in list(df_rename.columns):
        df_rename = df_rename.drop(columns = [6])
    if 7 in list(df_rename.columns):
        df_rename = df_rename.drop(columns = [7])
    
    df = pd.concat([df, df_rename])
```

```python
df[:-2]
```

```python
df[:-2].to_csv(f'{save_dir}/{os.path.basename(pdf_file).replace(".pdf",".csv")}')
```

## Converting all CSV

```python
for file in tqdm(pdf_files):
    if file == pdf_files[0]:
        continue
    dfs = read_pdf(file, pages = 'all')
    for j, df_tmp in enumerate(dfs):
        if j == 0:
            df = df_tmp.copy()
            continue
        df = pd.concat([df, df_tmp])

    df_rename = (
        df
        .rename(
            columns = {'Occupation of reporter\r(where known)' : 'Occupation of reporter (where known)'}
        )
    )
    df_rename.to_csv(f'{save_dir}/{os.path.basename(file).replace(".pdf",".csv")}')
```
